```python
import pandas as pd
import numpy as np

file_path = '/Users/aurorekouakou/Downloads/WA_Fn-UseC_-HR-Employee-Attrition.csv'
data = pd.read_csv(file_path)

print("Aperçu des données :")
print(data.head())

# Vérification des valeurs manquantes
print("\nValeurs manquantes par colonne :")
print(data.isnull().sum())

# Traitement des valeurs manquantes uniquement pour les colonnes numériques
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_cols] = data[numeric_cols].apply(lambda x: x.fillna(x.mean()))

# Transformation de certaines colonnes en catégories
for col in data.select_dtypes(include=['object']).columns:
    data[col] = data[col].astype('category')

# Affichage des types de données après transformation
print("\nTypes de données après transformation :")
print(data.dtypes)

# Sauvegarde des données nettoyées dans un fichier CSV
cleaned_file_path = 'cleaned_HR_data.csv'
data.to_csv(cleaned_file_path, index=False)
print(f"\nDonnées nettoyées sauvegardées dans {cleaned_file_path}")
```
